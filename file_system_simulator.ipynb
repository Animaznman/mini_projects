Simulate a tiny in‑memory file system.

Commands:

* `mkdir <path>`

* `touch <path>`

* `ls <path>`

* `tree` → print the entire directory tree

* `exit`

Paths look like:

Code
```bash
/home
/home/user
/home/user/docs
```

You must maintain a tree structure internally.

Questions I asked:
I know touch is a normal bash command, but I unfortunately don't remember what it does. Does it just try to return the shortest path to the desired path?

For printing the tree, is there a specific order the tree should be printed in?

For mkdir, what should the functionality be if there are collisions?

What is the format the tree should be printed in?
Great set of clarifying questions — these are exactly the things an interviewer expects you to ask before designing the internal tree structure. Let’s go through each one cleanly and precisely so you can build a correct, minimal, interview‑ready file system simulator.

⭐ 1. What does `touch <path>` do?
In Unix, touch:

creates an empty file if it doesn’t exist

updates the modification timestamp if it does

For this interview problem, you only need the first behavior:

✔ touch creates a file at the given path.
Example:

Code
`touch /home/user/docs/file.txt`
This means:

/home must exist

/home/user must exist

/home/user/docs must exist

then create a file named file.txt inside docs

If any parent directory does not exist, you have two choices:

Reject the command (most common in interviews)

Auto‑create missing directories (only if interviewer says so)

Default assumption: reject invalid paths.

⭐ 2. Is touch supposed to return the shortest path?
No.

touch does not return anything.
It simply creates a file node in your in‑memory tree.

⭐ 3. For printing the tree, is there a specific order?
Yes — and it’s simple:

✔ Directories first, then files, both sorted alphabetically.
This matches the behavior of tree on Linux.

Example:

Code
```
/home
    user
        docs
            file1.txt
            file2.txt
        pictures
            img.png
```
Indentation is typically:

4 spaces per level

or a tab per level

Either is acceptable unless the interviewer specifies otherwise.

⭐ 4. What should `mkdir <path>` do if there are collisions?
A “collision” means:

The directory already exists

A file exists with the same name

A parent component is a file instead of a directory

Here’s the expected behavior:

✔ Case 1 — Directory already exists
Do nothing (idempotent).

✔ Case 2 — A file exists with the same name
Error: cannot create a directory where a file exists.

✔ Case 3 — A parent component is a file
Error: cannot create a directory inside a file.

Example:

Code
touch /home/user/file
mkdir /home/user/file/docs
This should fail.

⭐ 5. What format should the tree be printed in?
Use a simple, readable, hierarchical format.

✔ Example format (recommended)
Code
```
/
    home/
        user/
            docs/
                file.txt
            pictures/
                img.png
```
Rules:

Directories end with /

Files do not

Indent by level

Alphabetical order

This is clean, predictable, and interview‑friendly.

````python
fs = {}

def mkdir(input_list):
    node = fs
    for part in input_list:
        if part not in node:
            node[part] = {}
        node = node[part]


mkdir(["home", "user", "docs"])

def check_dir(input_list):
    # Helper function to ensure the directories exist
    node = fs
    for part in input_list:
        if part in node:
            node = node[part]
        else:
            return False
    return True

def touch(input_list):
    if not check_dir(input_list[:-1]): # Ensure the directories exist up to the last item, which is the file that must be made.
        print("Invalid input.")
        return False
    node = fs
    for part in input_list[:-1]:
        node = node[part]
    node[input_list[-1]] = None

    return True

touch(["home", "user", "picture.png"])

def ls(input_list):
    if not check_dir(input_list): # Ensure the directories exist up to the last item, which is the file that must be made.
        print("Invalid input.")
        return False
    node = fs
    for part in input_list:
        node = node[part]
    print(node.keys())

mkdir(["home", "user1", "docs"])
touch(["home", "user1", "docs", "photo.png"])
ls(["home","user1"])

def tree_helper(node, depth):
    # node: current directory
    # depth: how far down the tree
    indent = "    " * (depth - 1)
    sorted_dir = [name for name in node if isinstance(node[name], dict)] # Check if dir is actually a directory by not being able to access its value
    sorted_dir.sort()

    sorted_files = [file for file in node if node[file] is None] # Opposite of above
    sorted_files.sort()

    for dir in sorted_dir:
        print(indent + dir + "/")
        tree_helper(node[dir], depth + 1)
    
    for file in sorted_files:
        print(indent + file)

def tree():
    print("/")
    tree_helper(fs, 1)
````
