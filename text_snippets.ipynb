Create a command‑line tool that allows a user to save, retrieve, list, and delete short text snippets. Each snippet should have a user‑provided name and a body of text. The tool should run interactively, accepting commands such as:
```bash
save greeting Hello there!
get greeting
list
delete greeting
```
Your task is to design and implement this tool.

Questions clarified by the interviewer:
1. For the get command, does it print to the screen, produce a txt file, or something else?
    
    For this version of the tool, get <name> should simply print the snippet’s text to the screen. No file creation is required. If you want to propose exporting to a file as an enhancement, that’s totally fine, but the base requirement is just terminal output.

2. Does it matter how the text snippet is saved? Should it be saved directly to the script, or something else?
    
    For the initial implementation, you can store snippets in memory — a dictionary or similar structure is fine. Persistence (saving to disk) is not required for the first pass. If you want to propose saving to a JSON file or similar for a second iteration, that’s a great enhancement, but not required.

3. For the list command, is there a limit to how much should be listed?
    
    No limit. list should display all snippet names currently stored. You don’t need pagination or truncation. Just print the names in whatever order you’re storing them — alphabetical ordering is optional but not required.

4. Is there an upper limit on the text you want to support for this program?
    
    There’s no strict upper bound. You can assume snippets will be reasonably short — a few sentences, maybe a paragraph — but your program doesn’t need to enforce a maximum length. If the user enters a very long snippet, your tool should still accept it.

````python
menu_message = """
Select a menu option: \n
1. save: Save a snippet \n
2. get: Get a snippet \n
3. list: List all saved snippets \n
4. delete: Delete a spcific snippet \n
5. exit: Exit the script \n
"""
user_input = int(input(menu_message))
snippets = {}
while user_input != 5:
    if user_input == 1:
        print("Saving a snippet.")
        snippet_key = input("Please provide a name for snippet. Keep the name short, 90 characters: ")
        while len(snippet_key) > 90: # On a future pass, do more validation stuff
            snippet_key = input("Improper length. Please provide a shorter snippet name: ")
        snippet_value = input("Please type in your actual snippet. Remember [ENTER] will input your message typed thus far: ")
        # Message validation in future pass
        snippets[snippet_key] = snippet_value
        print("Snippet saved!")
        user_input = int(input(menu_message))
    elif user_input == 2:
        print("Getting a snippet.")
        print("Please enter the number corresponding to the following saved snippets:")
        keys = snippets.keys()
        numeric_keys = {number: key for number, key in enumerate(keys)}
        for number, key in numeric_keys.items():
            print(f"{number}. {key}")
        target_snippet_number = int(input())
        print(snippets[numeric_keys[target_snippet_number]])
        print("Snippet returned!")
        user_input = int(input(menu_message))
    elif user_input == 3:
        print("Listing snippets.")
        print("Here are all the snippets locally saved: ")
        for index, (key, value) in enumerate(snippets.items()):
            if len(value) < 20: # Hard code check for seeing if the snippet is less than 20, future pass to make this variable
                print(f"{index}. {key}: {value}")
            else:
                value = value[:17]+"..."
                print(f"{index}. {key}: {value}")
        print("Snippets listed!")
        user_input = int(input(menu_message))
    elif user_input == 4:
        print("Deleting a snippet.")
        print("Please enter the number corresponding to the following saved snippets:")
        keys = snippets.keys()
        numeric_keys = {number: key for number, key in enumerate(keys)}
        for number, key in numeric_keys.items():
            print(f"{number}. {key}")
        target_snippet_number = int(input())
        # Future pass: add a message asking the user if they're sure they want to delete the snippet, while showing them a preview of it
        del snippets[numeric_keys[target_snippet_number]]
        print("Snippet deleted!")
        user_input = int(input(menu_message))
    elif user_input == 5:
        print("Now exiting the script. Thank you for using the snippets script!")
        user_input = 5
    else:
        print("Invalid selection.")
        user_input = int(input(menu_message))
````
