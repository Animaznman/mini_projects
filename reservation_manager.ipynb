# ⭐ 1. **Should reservation overlaps be allowed?**

**No. Overlapping reservations are not allowed.**

If John reserves **300–400**, and Joe tries to reserve **330–430**, the second reservation must **fail** because the intervals overlap.

This system supports **only one reservation at any given time**.

# ⭐ 2. **What should happen if `cancel` is given a different time span than the recorded reservation?**

### ✔ If the cancel interval exactly matches a reservation  
It is removed.

### ✔ If the cancel interval *does not* match any reservation  
It does **nothing** — no partial cancellation, no merging, no splitting.

### ✔ If the cancel interval partially overlaps a reservation  
It still does **nothing**.

Example:

- Existing: `reserve John 300 400`

Result:  
**No change.**  
The system does not split intervals.

### ✔ If the cancel interval fully contains a reservation  
Example:

- Existing: `reserve John 300 400`
- Command: `cancel John 200 500`

Result:  
**No change.**  
The system requires exact matching.

# ⭐ 3. **Can a user make overlapping reservations with themselves?**

**No.**

A user cannot reserve overlapping intervals with themselves or with anyone else.

Example:

- `reserve John 300 400` → OK  
- `reserve John 330 430` → FAIL  
- `reserve John 315 415` → FAIL  

### ✔ For `query <time>`  
You return **at most one name**, because only one reservation can exist at any given time.

# ⭐ 4. **For `schedule`, if two reservations have the same start time, which one comes first?**

This situation **cannot occur**, because overlapping reservations are disallowed.

If two reservations had the same start time, they would overlap, and the second one would have been rejected.

So the chronological order is always unambiguous.

````python
reservations = [] # Each resrvation is a tuple (name, start, end)

def reservation_overlaps(start, end, r_start, r_end):
    return not (end <= r_start or start >= r_end)

def reserve(name, start, end):
    for reservation in reservations:
        if reservation_overlaps(reservation[1], reservation[2], start, end):
            print("Error")
            return False
    
    reservations.append((name, start, end))
    reservations.sort(key = lambda x: x[1])
    print("Reservation complete")
    return True

def cancel(name, start, end):
    for i, (r_name, r_start, r_end) in enumerate(reservations):
        print((r_name, r_start, r_end))
        if r_name == name and r_start == start and r_end == end:
            reservations.pop(i)
            print("Reservation canceled")
            return True
    print("Reservation not found. Nothing cancelled.")
    return False

def query(time):
    for reservation in reservations:
        if reservation[1] <= time < reservation[2]:
            print(f"Reservation at time {time}.")
            print(f"{reservation[0]} from {reservation[1]} until {reservation[2]}.")
            return True
    
    print(f"No reservations found at time {time}")
    return False

def schedule():
    for reservation in reservations:
        print(f"{reservation[0]} from {reservation[1]} until {reservation[2]}.")
````
