Create a CLI tool that manages meeting requests. Each meeting is:

Code
`<start_time> <end_time>`

Commands:

* `add <start> <end>` → add a meeting

* `conflicts` → list all pairs of meetings that overlap

* `free <start> <end>` → list all free time blocks within the range

* `exit`

```python
meetings = []

def find_first_index_next(data_list, condition_func, default=None):
  """Finds the index of the first element meeting the condition using next()."""
  # Generator yields index if condition is met
  generator = (index for index, item in enumerate(data_list) if condition_func(item))
  # next() retrieves the first item from the generator
  # default is returned if the generator is empty (no match)
  return next(generator, default)

def add(start, end):
    meetings.append((start, end))

def conflicts():
    sorted_meetings = sorted(meetings, key= lambda x: x[0])
    conflicting_meetings = []
    for index in range(len(sorted_meetings)-1):
        cur_start, cur_end = sorted_meetings[index]

        relative_index = find_first_index_next(sorted_meetings[index + 1:], lambda x: x[0] > cur_end, default = None)

        if relative_index is None:
            first_open_index = len(sorted_meetings)
        else: first_open_index = index + 1 + relative_index

        for sub_index in range(index + 1, first_open_index):
            next_start, next_end = sorted_meetings[sub_index]

            if next_start < cur_end:
                conflicting_meetings.append((sorted_meetings[index], sorted_meetings[sub_index]))

    
    return conflicting_meetings

add(100, 200)
add(201, 350)
add(300, 400)

conflicts()

def free(start, end):
    conflicting_times = []
    for meeting in meetings:
        start_time, end_time = meeting
        if end_time > start or start_time < end:
            conflicting_times.append(meeting)
    
    conflicting_times.sort()
    free_times = []
    cur_time = start
    for s, e in conflicting_times:
        if s > cur_time:
            free_times.append((cur_time, s))
        cur_time = max(cur_time, e)
    
    if cur_time < end:
        free_times.append((cur_time, end))
    
    return free_times

free(0, 500)
```
