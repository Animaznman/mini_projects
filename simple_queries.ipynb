Create a command‑line tool that loads a CSV file and allows the user to run simple queries like:
`filter age > 30` or `select name,city`.

* `filter` - display only results that match the logic following the command
* `select` - display the columns of the listed values

`select city, occupation, filter salary > 100000 order by city DESC save 'output.csv'`

First pass:
`python queries.py`

Use pandas to load csv
user_input = input("Please input a command") # Take whatever the user has as an input

While user_input != 'exit'
1. parse user_input to see if 'filter' -> do filter stuff
2. parse `... select` -> do select stuff
3. parse `... exit` -> Friendly exit message -> exit script
4. None of the above -> Reminder to use keywords -> back to top of while loop

```python
import pandas as pd

dataframe = pd.read_csv('people.csv')
data = [dataframe.columns.tolist()] # Future: Possibly use the headers as a dictionary and the value as the index
data.extend(dataframe.values.tolist())

# Future: Valid inputs dictionary/set for future attempts
valid_comparisons = ("<", ">", "=")

user_input = input("Please input a proper command: ").lower() # filter should be the last part of the input
parsed_input = user_input.split(" ") # Each column should have , after it for select statement
dummy_input = "select city, occupation, filter salary > 100000"
while 'exit' not in parsed_input:
    if 'filter' in parsed_input:
        # Do the filtering stuff
        filter_parse = parsed_input[parsed_input.index('filter') + 1:] # Get all list elements from 'filter on
        if filter_parse[0] in data[0]:
            value = eval(filter_parse[-1]) # Change the string value to an actual numeric value
            if filter_parse[1] == '>':
                # do greater than stuff
                column_index = data[0].index(filter_parse[0]) # Index of the target column
                final_data = [datum for datum in data[1:] if datum[column_index] > value]
                # print(final_data) # debug
                # Future: Test the input once you've done some major aspect/function
            elif filter_parse[1] == '<':
                # less than stuff
                continue 
            elif filter_parse[1] == '=':
                # do equal to stuff
                continue
            else:
                # exit/error out
                continue
        else: # filtered word not existant, exit/error out
            continue
    if 'select' in parsed_input:
        target_columns = [column for column in parsed_input[1:] if ',' in column]
        parsed_columns = [column[:-1] for column in target_columns]
        target_indices = [data[0].index(column) for column in parsed_columns]
        final_data = [[row[target_index] for target_index in target_indices] for row in final_data]
    print(final_data)
    if 'exit' in parsed_input:
        print("Thank you for using this script. Bye bye.")
        break
    user_input = input("Please input a proper command: ").lower()
    parsed_input = user_input.split(" ")
```
